### Hello World Example of ML

In this notebook you will serve your first TensorFlow model with TensorFlow Serving. We will start by building a very simple model to infer the relationship:

$$
y = 3x + 1 
$$

[source](https://codelabs.developers.google.com/codelabs/tensorflow-lab1-helloworld/)

```python
import os

import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```python
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)
```

```python
model.fit(xs, ys, epochs=500)
```

```python
print(model.predict([10.0]))
```

```python
MODEL_DIR = 'helloworld'

version = 1

export_path = os.path.join(MODEL_DIR, str(version))

tf.keras.models.save_model(model, export_path)

print('\nexport_path = {}'.format(export_path))
!ls -l {export_path}
```

## Inference With Rest


```python
import json
import requests
```

```python
json_response = requests.get('http://localhost:8501/v1/models/helloworld/metadata')
print(json_response.text)
```

```python
xs = np.array([[9.0], [10.0]])
data = json.dumps({"signature_name": "serving_default", "inputs": xs.tolist()})
print(data)
```

```python
json_response = requests.get('http://localhost:8501/v1/models/helloworld/metadata')
print(json_response.text)
```

```python
headers = {"content-type": "application/json"}
json_response = requests.post('http://localhost:8501/v1/models/helloworld:predict', data=data, headers=headers)

print(json_response.text)
```
